```python
import numpy as np
import matplotlib.pyplot as plt
```

# Latenz

### Skalierende N_Trees bei Max_Depth = 8

```python
x = np.array([1,8,16,32,64,128,256,512])

y_sklearn  = np.array([0.22, 0.66, 1.13, 2.05, 3.95, 7.70, 15.02, 29.76])
y_gemm_cpu = np.array([0.04, 0.07, 0.09, 0.13, 0.20, 0.36, 1.08, 2.35])
y_tt_cpu   = np.array([0.11, 0.10, 0.11, 0.11, 0.11, 0.12, 0.13, 0.14])
y_ptt_cpu  = np.array([0.08, 0.10, 0.10, 0.10, 0.10, 0.11, 0.11, 0.11])

y_gemm_ncs = np.array([0.39, 0.59, 1.24, 2.10, 4.52, 7.58, 6.64, 12.56])
y_tt_ncs   = np.array([2.55, 3.75, 4.04, 2.98, 3.22, 4.38, 5.79, 8.58])
y_ptt_ncs  = np.array([2.68, 3.61, 3.35, 2.45, 2.70, 3.20, 4.18, 6.73])
```

```python
fig, ax = plt.subplots()

ax.set_axisbelow(True)
ax.grid()
ax.plot(x, y_sklearn, c='red')
ax.plot(x, y_ptt_ncs, c='green')
ax.plot(x, y_ptt_cpu, c='blue')
plt.legend(labels=['Sklearn', 'PTT auf NCS2', 'PTT auf CPU'])
ax.scatter(x, y_ptt_ncs, c='green', marker='s')
ax.scatter(x, y_ptt_cpu, c='blue', marker='s')
ax.scatter(x, y_sklearn, c='red', marker='s')
plt.xlabel("Anzahl der BÃ¤ume")
plt.ylabel("Inferenzzeit in ms")
```

```python
fig, ax = plt.subplots()

plt.title("HB-Methoden auf CPU")
ax.set_axisbelow(True)
ax.grid()
ax.plot(x, y_gemm_cpu, c='red')
ax.plot(x, y_tt_cpu, c='green')
ax.plot(x, y_ptt_cpu, c='blue')
plt.legend(labels=['GEMM', 'TT', 'PTT'])
ax.scatter(x, y_ptt_cpu, c='blue', marker='s')
ax.scatter(x, y_tt_cpu, c='green', marker='s')
ax.scatter(x, y_gemm_cpu, c='red', marker='s')
plt.xlabel("Anzahl der BÃ¤ume")
plt.ylabel("Inferenzzeit in ms")
```

```python
fig, ax = plt.subplots()

plt.title("HB-Methoden auf NCS2")
ax.set_axisbelow(True)
ax.grid()
ax.plot(x, y_gemm_ncs, c='red')
ax.plot(x,   y_tt_ncs, c='green')
ax.plot(x,  y_ptt_ncs, c='blue')
plt.legend(labels=['GEMM', 'TT', 'PTT'])
ax.scatter(x,  y_ptt_ncs, c='blue', marker='s')
ax.scatter(x,   y_tt_ncs, c='green', marker='s')
ax.scatter(x, y_gemm_ncs, c='red', marker='s')
plt.xlabel("Anzahl der BÃ¤ume")
plt.ylabel("Inferenzzeit in ms")
```
